```python
import json
from collections import defaultdict

import numpy as np
import pandas

import matplotlib.pyplot as plt
from matplotlib.collections import PatchCollection

from mpl_toolkits.basemap import Basemap

from shapely.geometry import Polygon, MultiPolygon
from descartes import PolygonPatch
```

```python
PATH = "/Users/dmasad/Programming/workspace/MasonProjects/"
data = json.load(open(PATH + "RiskOut.json"))
```

```python
m = Basemap(projection='robin', lon_0=0)
m.readshapefile(PATH + "src/riskworld/data/world_country_boundary_file_with_fips", 'world', color='none', zorder=2)

def draw_map(country_data, colorbar_legend=""):
    df_map = pandas.DataFrame({"polygon": [Polygon(xy) for xy in m.world],
                           "Name": [shape["NAME_SORT"] for shape in m.world_info]})
    df_map["patches"] = df_map.polygon.apply(lambda x: 
                    PolygonPatch(x, fc='#555555', ec='#787878', lw=0.25, alpha=1, zorder=4))
    
    country_df = pandas.DataFrame({"Data": country_data})
    df_map = df_map.merge(country_df, how='left', left_on="Name", right_index=True)
    df_map.Data.fillna(0, inplace=True)
    cmap = plt.get_cmap("Reds")
    pc = PatchCollection(df_map["patches"].values, match_original=True)
    pc.set_facecolor(cmap(df_map.Data.values))

    fig = plt.figure(figsize=(32,12))
    ax = fig.add_subplot(111)
    m.drawmapboundary(fill_color= '#eeeeee')
    ax.add_collection(pc)
    
    mappable = cm.ScalarMappable(cmap=cmap)
    mappable.set_array(df_map.Data.unique())
    cb = fig.colorbar(mappable)
    cb.set_label(colorbar_legend)
    return (fig, ax)
    
```

## Supply security

```python
# Chance of a major supply shock:
threshold = 2
country_ratios = defaultdict(list)
for iteration in data:
    for country, supply_ratios in iteration["supplyRatios"].iteritems():
        country_ratios[country] += supply_ratios
country_risk = {}
for country, ratios in country_ratios.iteritems():
    r = np.array(ratios)
    country_risk[country] = len(r[r>threshold])*1.0/len(r)

fig, ax = draw_map(country_risk, "Chance of major supply shock")
```

## Demand security

```python
# Compute chance of a major (<0.5) demand shock:
threshold = 0.5
country_ratios = defaultdict(list)
for iteration in data:
    for country, supply_ratios in iteration["demandRatios"].iteritems():
        country_ratios[country] += supply_ratios
country_risk = {}
for country, ratios in country_ratios.iteritems():
    r = np.array(ratios)
    country_risk[country] = len(r[r<threshold])*1.0/len(r)

# Remove bad data
for key, val in country_risk.iteritems():
    if val == 1:
        country_risk[key] = 0
    
draw_map(country_risk, "Chance of major demand shock")
```
