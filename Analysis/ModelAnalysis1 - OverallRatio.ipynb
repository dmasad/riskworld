```python
import pandas
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
import os
import json
import numpy as np
import scipy.stats
import math
from collections import defaultdict
from itertools import cycle
```

```python
data = json.load(open(PATH + "RiskOut.json"))
scenario_data = defaultdict(list)
for entry in data:
    scenario = (entry['contagion'], entry['assistance'])
    scenario_data[scenario].append(entry)
```

## Contagion vs non-contagion

```python
# Contagion vs non-contagion
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111, axis_bgcolor="#eeeeee")
ax.grid(True)
all_ratios = defaultdict(list)

for scenario, iterations in scenario_data.iteritems():
    realizations = [x["globalOverallRatio"] for x in iterations]
    for realization in realizations:
        all_ratios[scenario[0]] += realization

h = ax.hist(all_ratios[False], bins=np.linspace(0,3,31), color='cyan', alpha=0.75)
h = ax.hist(all_ratios[True], bins=np.linspace(0,3,31), color='r', alpha=0.75)
```

```python
# Contagion vs non-contagion
fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot(111, axis_bgcolor="#eeeeee")
ax.grid(True)
all_ratios = defaultdict(list)

for scenario, iterations in scenario_data.iteritems():
    realizations = [x["globalOverallRatio"] for x in iterations]
    for realization in realizations:
        all_ratios[scenario[0]] += realization

h = sns.kdeplot(all_ratios[False], shade=True, label="No Contagion")
h = sns.kdeplot(all_ratios[True], shade=True, label="Contagion")
plt.legend(fontsize=28)

plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.xlabel("Global Ratio", fontsize=24)
plt.ylabel("PDF", fontsize=24)
```

```python
false_data = np.array(all_ratios[False])
true_data = np.array(all_ratios[True])
# Comparison
print "Data Summary"
print "====================="
print "Parameter = False summary:"
print scipy.stats.describe(false_data)
print "Parameter = True:"
print scipy.stats.describe(true_data)
print "\nAre the data drawn from the same distribution?"
print "p-vals:"
print "K-S test: ", scipy.stats.ks_2samp(false_data, true_data)[1]
print "Mann-Whitney:", scipy.stats.mannwhitneyu(false_data, true_data)[1]
print "Are the means identical?"
print "t-test:", scipy.stats.ttest_ind(false_data, true_data, equal_var=False)[1]
print "Wilcoxon:", scipy.stats.wilcoxon(false_data, true_data)[1]
```

```python
# Summarize effects of difference
print "Mean effect magnitude", (np.mean(true_data)-np.mean(false_data))/np.mean(false_data)
print "Variance effect magnitude", (np.var(true_data)-np.var(false_data))/np.var(false_data)
print "Probabilities of serious crisis:"
print "Without contagion", 1.0*len(false_data[false_data>1.5])/len(false_data)
print "With contagion", 1.0*len(true_data[true_data>1.5])/len(true_data)
```

```python
#Bayesian confidence interval
print scipy.stats.bayes_mvs(false_data)[0] 
print scipy.stats.bayes_mvs(true_data)[0]
```

```python
# Non-Bayesian Confidence Interval
def mean_confidence_interval(data, confidence=0.95):
    a = 1.0*np.array(data)
    n = len(a)
    m, se = np.mean(a), scipy.stats.sem(a)
    h = se * scipy.stats.t._ppf((1+confidence)/2., n-1)
    return m, m-h, m+h

print mean_confidence_interval(false_data)
print mean_confidence_interval(true_data)
```

## Correlation

```python
means = []
varis = []

for scenario, iterations in scenario_data.iteritems():
    realizations = [x["globalOverallRatio"] for x in iterations]
    means += [np.mean(x) for x in realizations]
    varis += [np.var(x) for x in realizations]

fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot(111, axis_bgcolor="#eeeeee")
ax.scatter(varis, means, c='k', s=20, lw=0) # prev: c="#666666"
ax.set_xlabel("Variance", fontsize=24)
ax.set_ylabel("Mean", fontsize=24)
ax.grid(True)
ax.set_xlim(0,0.1)
ax.set_xticklabels(ax.get_xticks(), fontsize=16)
ax.set_yticklabels(ax.get_yticks(), fontsize=16)
```
