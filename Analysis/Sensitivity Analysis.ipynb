## Overall Ratio

```python
# Overall Ratio
fig = plt.figure(figsize=(12,6))

# Means
pivot = pandas.pivot_table(model_df, values="globalOverallRatio", 
                   rows="conflictExponent", cols="defaultSpareCapacity")
ax1 = fig.add_subplot(121)
pcol = ax1.pcolor(pivot)
ax1.set_yticks(arange(0.5, len(pivot.index), 1))
ax1.set_yticklabels(pivot.index)

ax1.set_xticks(arange(0.5, len(pivot.columns), 1))
ax1.set_xlim(0, len(pivot.columns))
ax1.set_xticklabels(pivot.columns)
fig.colorbar(pcol, ax=ax1)

# Variance
pivot = pandas.pivot_table(model_df, values="globalOverallRatio", 
                   rows="conflictExponent", cols="defaultSpareCapacity", aggfunc=np.var)
ax2 = fig.add_subplot(122)
pcol = ax2.pcolor(pivot)
ax2.set_yticks(arange(0.5, len(pivot.index), 1))
ax2.set_yticklabels(pivot.index)

ax2.set_xticks(arange(0.5, len(pivot.columns), 1))
ax2.set_xlim(0, len(pivot.columns))
ax2.set_xticklabels(pivot.columns)
fig.colorbar(pcol, ax=ax2)

# Labeling
ax1.set_title("Overall Mean")
ax2.set_title("Overall Variance")
ax1.set_xlabel("Default Spare Capacity")
ax2.set_xlabel("Default Spare Capacity")
ax1.set_ylabel("Conflict Exponent")
ax2.set_ylabel("Conflict Exponent")
```
